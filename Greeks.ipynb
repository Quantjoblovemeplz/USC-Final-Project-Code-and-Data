```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for Black-Scholes Delta
S = np.linspace(75, 125, 400)  # Spot price range
K = 100  # Strike price
r = 0.01  # Risk-free rate
sigma = 0.2  # Volatility

# Black-Scholes Delta formula for call option
def black_scholes_delta_call(S, K, t, r, sigma):
    d1 = (np.log(S / K) + (r + 0.5 * sigma ** 2) * t) / (sigma * np.sqrt(t))
    delta_call = norm.cdf(d1)
    return delta_call

# Calculate Delta for different times to maturity
tau_10_days = black_scholes_delta_call(S, K, 10/365, r, sigma)
tau_1_month = black_scholes_delta_call(S, K, 1/12, r, sigma)
tau_3_months = black_scholes_delta_call(S, K, 3/12, r, sigma)

# Plotting the Deltas
plt.figure(figsize=(10, 6))
plt.plot(S, tau_10_days, label=r'$\tau = 10 days$')
plt.plot(S, tau_1_month, '--', label=r'$\tau = 1 month$')
plt.plot(S, tau_3_months, ':', label=r'$\tau = 3 months$')

plt.title('Delta of Call Options')
plt.xlabel('Spot Price')
plt.ylabel('Delta')
plt.legend()
plt.grid(True)
plt.show()
```

```python
S = np.linspace(75, 125, 400)  # Spot price range
K = 100  # Strike price
r = 0.01  # Risk-free rate
sigma = 0.2  # Volatility

# Black-Scholes Gamma formula for both call and put option
def black_scholes_gamma(S, K, t, r, sigma):
    d1 = (np.log(S / K) + (r + 0.5 * sigma ** 2) * t) / (sigma * np.sqrt(t))
    gamma = norm.pdf(d1) / (S * sigma * np.sqrt(t))
    return gamma

# Calculate Gamma for different times to maturity
tau_10_days = black_scholes_gamma(S, K, 10/365, r, sigma)
tau_1_month = black_scholes_gamma(S, K, 1/12, r, sigma)
tau_3_months = black_scholes_gamma(S, K, 3/12, r, sigma)

# Plotting the Gammas
plt.figure(figsize=(10, 6))
plt.plot(S, tau_10_days, label=r'$\tau = 10 days$')
plt.plot(S, tau_1_month, '--', label=r'$\tau = 1 month$')
plt.plot(S, tau_3_months, ':', label=r'$\tau = 3 months$')

plt.title('Gamma of Call/Put Options')
plt.xlabel('Spot Price')
plt.ylabel('Gamma')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Parameters for Black-Scholes Theta
S = np.linspace(75, 125, 400)  # Spot price
K = 100  # Strike price
r = 0.01  # Risk-free rate
sigma = 0.2  # Volatility

# Black-Scholes Theta formula for call options
def black_scholes_theta_call(S, K, t, r, sigma):
    d1 = (np.log(S / K) + (r + 0.5 * sigma ** 2) * t) / (sigma * np.sqrt(t))
    d2 = d1 - sigma * np.sqrt(t)
    theta = -(S * sigma * norm.pdf(d1)) / (2 * np.sqrt(t)) - r * K * np.exp(-r * t) * norm.cdf(d2)
    return theta / 365  # Convert to per day

# Calculate Theta for different times to maturity
theta_10_days = black_scholes_theta_call(S, K, 10/365, r, sigma)
theta_1_month = black_scholes_theta_call(S, K, 1/12, r, sigma)
theta_3_months = black_scholes_theta_call(S, K, 3/12, r, sigma)

# Plotting the Thetas
plt.figure(figsize=(10, 6))
plt.plot(S, theta_10_days, label=r'$\tau = 10 days$')
plt.plot(S, theta_1_month, '--', label=r'$\tau = 1 month$')
plt.plot(S, theta_3_months, ':', label=r'$\tau = 3 months$')

plt.title('Theta of Call Options')
plt.xlabel('Spot Price')
plt.ylabel('Theta')
plt.legend()
plt.grid(True)
plt.show()
```
