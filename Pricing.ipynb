```python
import numpy as np 
import pandas as pd
import math
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
e = math.e
# Black-Scholes model pricing
def price(rf, σ, t, K, S):
    d_1 = (math.log(S / K) + (rf + σ**2 / 2) * t) / (σ * math.sqrt(t))
    d_2 = d_1 - σ * math.sqrt(t)
    Call_price = S * norm.cdf(d_1) - K * e**(-rf * t) * norm.cdf(d_2)
    return Call_price
```
